### Get List & Detail of Parqueat Files

```python
import os
import pyarrow.parquet as pq
import pandas as pd

def list_parquet_files(base_dir):
    parquet_info = []

    for root, _, files in os.walk(base_dir):
        for file in files:
            if file.endswith(".parquet"):
                file_path = os.path.join(root, file)
                try:
                    # Get file size (bytes)
                    file_size = os.path.getsize(file_path)
                    
                    # Read only Parquet metadata (fast, no data load)
                    metadata = pq.ParquetFile(file_path).metadata
                    num_rows = metadata.num_rows

                    parquet_info.append({
                        "file_name": file,
                        "file_path": file_path,
                        "size_MB": round(file_size / (1024 * 1024), 2),
                        "num_rows": num_rows
                    })
                except Exception as e:
                    print(f"âš ï¸ Error reading {file_path}: {e}")

    # Sort by file name
    parquet_info.sort(key=lambda x: x["file_name"])

    # Convert to DataFrame for clean display
    df = pd.DataFrame(parquet_info)

    # Print results
    print(df.to_string(index=False))

    # Show total size
    total_size = sum(f["size_MB"] for f in parquet_info)
    print("\n" + "-" * 80)
    print(f"Total size: {total_size:.2f} MB across {len(parquet_info)} files")

# Example usage:
list_parquet_files("/home/dev/workstation/data")
```

```python
import pandas as pd

# --- File paths ---
quotes_path = "data/quote/2025-10-13/crypto_quotes_2025-10-13_03-16.parquet"
trades_path = "data/trade/2025-10-13/crypto_trades_2025-10-13_03-16.parquet"

# --- Read Parquet files ---
df_quotes = pd.read_parquet(quotes_path)
df_trades = pd.read_parquet(trades_path)

# --- Convert timestamp to datetime ---
df_quotes['datetime'] = pd.to_datetime(df_quotes['timestamp'], unit='ms')
df_trades['datetime'] = pd.to_datetime(df_trades['timestamp'], unit='ms')

# --- Count before filtering ---
quotes_before = len(df_quotes)
trades_before = len(df_trades)

# --- Add datetime columns ---
for df in [df_quotes, df_trades]:
    df['year'] = df['datetime'].dt.year
    df['month'] = df['datetime'].dt.month
    df['day'] = df['datetime'].dt.day
    df['hour'] = df['datetime'].dt.hour
    df['minutes'] = df['datetime'].dt.minute
    df['sec'] = df['datetime'].dt.second
    df['ms'] = df['datetime'].dt.microsecond // 1000

# --- Filter Ticker & Date-Time ---
# --- Filter condition ---
filter_quotes = (
    (df_quotes['pair'] == 'BTC-USD')
    & (df_quotes['year'] == 2025) 
    & (df_quotes['month'] == 10) 
    & (df_quotes['day'] == 13) 
    & (df_quotes['hour'] == 3) 
    & (df_quotes['minutes'] == 16) 
    #& (df_quotes['sec'] == 4)
)
filter_trades = (
    (df_trades['pair'] == 'BTC-USD') 
    & (df_trades['year'] == 2025) 
    & (df_trades['month'] == 10) 
    & (df_trades['day'] == 13) 
    & (df_trades['hour'] == 3) 
    & (df_trades['minutes'] == 16) 
    #& (df_trades['sec'] == 4)
)

# --- Apply filters ---
quotes_filtered = df_quotes.loc[filter_quotes]
trades_filtered = df_trades.loc[filter_trades]

# --- Count after filtering ---
quotes_after = len(quotes_filtered)
trades_after = len(trades_filtered)

# --- Show summary ---
print("ðŸ“Š Row Count Summary")
print(f"Quotes: before={quotes_before:,} â†’ after={quotes_after:,}")
print(f"Trades: before={trades_before:,} â†’ after={trades_after:,}")
```

```python
#â€œWithin each second, how many distinct price changes occur?â€

# --- Detect price changes (True where current != previous) ---
df_quotes['price_changed'] = quotes_filtered['bid_price'].ne(quotes_filtered['bid_price'].shift())
df_trades['price_changed'] = trades_filtered['price'].ne(trades_filtered['price'].shift())

# --- Count price changes per second ---
quotes_price_changes = (
    df_quotes.groupby('sec')['price_changed']
    .sum()
    .reindex(range(0, 60), fill_value=0)
    .astype(int)
)

trades_price_changes = (
    df_trades.groupby('sec')['price_changed']
    .sum()
    .reindex(range(0, 60), fill_value=0)
    .astype(int)
)

# --- Combine into single summary DataFrame ---
df_price_change_summary = pd.DataFrame({
    'sec': range(0, 60),
    'quotes_price_changes': quotes_price_changes.values,
    'trades_price_changes': trades_price_changes.values
})

# --- Show results ---
print("ðŸ“Š Price Change Frequency per Second (0â€“59):")
print(df_price_change_summary)
```

```python
import base64
import hashlib
import hmac
import json
import logging
import os
import time
import uuid
from urllib.parse import urlencode
import requests


class KcSigner:
    def __init__(self, api_key: str, api_secret: str, api_passphrase: str):
        """
        KcSigner contains information about `apiKey`, `apiSecret`, `apiPassPhrase`
        and provides methods to sign and generate headers for API requests.
        """
        self.api_key = api_key or ""
        self.api_secret = api_secret or ""
        self.api_passphrase = api_passphrase or ""

        if api_passphrase and api_secret:
            self.api_passphrase = self.sign(api_passphrase.encode('utf-8'), api_secret.encode('utf-8'))

        if not all([api_key, api_secret, api_passphrase]):
            logging.warning("API token is empty. Access is restricted to public interfaces only.")

    def sign(self, plain: bytes, key: bytes) -> str:
        hm = hmac.new(key, plain, hashlib.sha256)
        return base64.b64encode(hm.digest()).decode()

    def headers(self, plain: str) -> dict:
        """
        Headers method generates and returns a map of signature headers needed for API authorization
        It takes a plain string as an argument to help form the signature. The outputs are set of API headers.
        """
        timestamp = str(int(time.time() * 1000))
        signature = self.sign((timestamp + plain).encode('utf-8'), self.api_secret.encode('utf-8'))

        return {
            "KC-API-KEY": self.api_key,
            "KC-API-PASSPHRASE": self.api_passphrase,
            "KC-API-TIMESTAMP": timestamp,
            "KC-API-SIGN": signature,
            "KC-API-KEY-VERSION": "3"
        }


def process_headers(signer: KcSigner, body: bytes, raw_url: str, request: requests.PreparedRequest, method: str):
    request.headers["Content-Type"] = "application/json"

    # Create the payload by combining method, raw URL, and body
    payload = method + raw_url + body.decode()
    headers = signer.headers(payload)

    # Add headers to the request
    request.headers.update(headers)


def get_trade_fees(signer: KcSigner, session: requests.Session):
    endpoint = "https://api.kucoin.com"
    path = "/api/v1/trade-fees"
    method = "GET"
    query_params = {"symbols": "BTC-USDT"}

    # Build full URL and raw URL
    full_path = f"{endpoint}{path}?{urlencode(query_params)}"
    raw_url = f"{path}?{urlencode(query_params)}"

    req = requests.Request(method=method, url=full_path).prepare()
    process_headers(signer, b"", raw_url, req, method)

    resp = session.send(req)
    resp_obj = json.loads(resp.content)
    print(resp_obj)

def get_all_tickers(signer: KcSigner, session: requests.Session):
    endpoint = "https://api.kucoin.com"
    path = "/api/v1/market/allTickers"
    method = "GET"
    query_params = {"symbols": "BTC-USDT"}
    

    # Build full URL and raw URL
    full_path = f"{endpoint}{path}?{urlencode(query_params)}"
    raw_url = f"{path}?{urlencode(query_params)}"

    req = requests.Request(method=method, url=full_path).prepare()
    process_headers(signer, b"", raw_url, req, method)

    resp = session.send(req)
    resp_obj = json.loads(resp.content)
    print(resp_obj)



def add_limit_order(signer: KcSigner, session: requests.Session):
    endpoint = "https://api.kucoin.com"
    path = "/api/v1/hf/orders"
    method = "POST"

    # Prepare order data
    order_data = json.dumps({
        "clientOid": str(uuid.uuid4()),
        "side": "buy",
        "symbol": "BTC-USDT",
        "type": "limit",
        "price": "10000",
        "size": "0.001",
    })

    # Build full URL and raw URL
    full_path = f"{endpoint}{path}"
    raw_url = path

    req = requests.Request(method=method, url=full_path, data=order_data).prepare()
    process_headers(signer, order_data.encode(), raw_url, req, method)

    resp = session.send(req)
    resp_obj = json.loads(resp.content)
    print(resp_obj)


if __name__ == '__main__':
    # Load API credentials from environment variables
    key = os.getenv("API_KEY", "68f3150503ad1c00011b1fd0")
    secret = os.getenv("API_SECRET", "9977fb96-6226-4131-9716-43d3f191b373")
    passphrase = os.getenv("API_PASSPHRASE", "kucoin_api")

    session = requests.Session()
    signer = KcSigner(key, secret, passphrase)

    # Perform operations
    #get_trade_fees(signer, session)
    get_all_tickers(signer, session)
    #add_limit_order(signer, session)
```

```python
from datetime import datetime, timezone, timedelta

def get_rounded_timestamps(interval_minutes=15):
    # current UTC time (timezone-aware)
    now_utc = datetime.now(timezone.utc)
    
    # round down to nearest interval (e.g., 15 minutes)
    rounded_minute = (now_utc.minute // interval_minutes) * interval_minutes
    rounded_utc = now_utc.replace(minute=rounded_minute, second=0, microsecond=0)
    
    # convert UTC to Eastern Time (UTC-5)
    ET_OFFSET = timedelta(hours=-5)
    rounded_et = rounded_utc.astimezone(timezone(ET_OFFSET))
    
    # timestamps
    utc_unix = int(rounded_utc.timestamp())
    et_unix = int(rounded_et.timestamp())
    
    return {
        "utc_time": rounded_utc.strftime("%Y-%m-%d %H:%M:%S %Z"),
        "utc_unix": utc_unix,
        "et_time": rounded_et.strftime("%Y-%m-%d %I:%M:%S %p %Z"),
        "et_unix": et_unix
    }
```

```python
import os
import json
import pyarrow.parquet as pq
import pandas as pd
from collections import defaultdict

# === Helper: flatten nested JSON ===
def flatten_keys(obj, prefix=""):
    keys = set()
    if isinstance(obj, dict):
        for k, v in obj.items():
            full_key = f"{prefix}.{k}" if prefix else k
            keys.add(full_key)
            keys |= flatten_keys(v, full_key)
    elif isinstance(obj, list):
        for item in obj:
            keys |= flatten_keys(item, prefix)
    return keys

# === Helper: get schema of a single JSON string ===
def get_json_schema(row_value):
    try:
        data = json.loads(row_value)
        return tuple(sorted(flatten_keys(data)))
    except Exception:
        return None

# === Main function: extract schemas from all parquet files ===
def extract_unique_schemas(folder_path):
    schema_set = set()
    schema_examples = defaultdict(list)

    for file in os.listdir(folder_path):
        if not file.endswith(".parquet"):
            continue

        filepath = os.path.join(folder_path, file)
        try:
            table = pq.read_table(filepath, columns=["raw"])
        except Exception as e:
            print(f"âš ï¸ Error reading {file}: {e}")
            continue

        for value in table.column("raw").to_pylist():
            schema = get_json_schema(value)
            if schema and schema not in schema_set:
                schema_set.add(schema)
                schema_examples[schema].append(file)

    return schema_set, schema_examples

# === Run analysis ===
folder_path = "./"  # ðŸ”§ change to your folder path

unique_schemas, schema_examples = extract_unique_schemas(folder_path)

print(f"âœ… Found {len(unique_schemas)} unique JSON schemas\n")

for i, schema in enumerate(unique_schemas, 1):
    print(f"Schema #{i}: {len(schema)} fields")
    for path in schema:
        print("   ", path)
    print("-" * 60)

# === Optional: export to CSV for later review ===
schema_records = [
    {"schema_id": i + 1, "field_path": path}
    for i, schema in enumerate(unique_schemas)
    for path in schema
]

if schema_records:
    df = pd.DataFrame(schema_records)
    df.to_csv("unique_json_schemas.csv", index=False)
    print("\nðŸ“ Saved schema summary â†’ unique_json_schemas.csv")
else:
    print("âš ï¸ No schemas found or JSONs invalid.")
```

### FN :: get_rounded_timestamps

```python
from datetime import datetime, timezone, timedelta

def get_rounded_timestamps(interval_minutes=15):
    # current UTC time (timezone-aware)
    now_utc = datetime.now(timezone.utc)
    
    # round down to nearest interval (e.g., 15 minutes)
    rounded_minute = (now_utc.minute // interval_minutes) * interval_minutes
    rounded_utc = now_utc.replace(minute=rounded_minute, second=0, microsecond=0)
    
    # convert UTC to Eastern Time (UTC-5)
    ET_OFFSET = timedelta(hours=-5)
    rounded_et = rounded_utc.astimezone(timezone(ET_OFFSET))
    
    # timestamps
    utc_unix = int(rounded_utc.timestamp())
    et_unix = int(rounded_et.timestamp())
    
    return {
        "utc_time": rounded_utc.strftime("%Y-%m-%d %H:%M:%S %Z"),
        "utc_unix": utc_unix,
        "et_time": rounded_et.strftime("%Y-%m-%d %I:%M:%S %p %Z"),
        "et_unix": et_unix
    }

# Example use:
times = get_rounded_timestamps()
print(times["utc_time"])
print(times["utc_unix"])
print(times["et_time"])
print(times["et_unix"])

print("Curent Market BTC 15min:", f"https://polymarket.com/event/btc-updown-15m-{times["utc_unix"]}")

```
